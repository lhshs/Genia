### 1. Read Module

```python
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.graph_objects as go
from dash import dash_table as dt
from dash import State

from datetime import datetime
import pandas as pd
import numpy as np
```

### 2. Read CSV File

```python
df_company_info = pd.read_csv('data/df_krx.csv', index_col=0)
# 6자리 숫자 맞춰주기, 앞에 0을 붙여주기 위해 map을 사용
df_company_info['Symbol'] = df_company_info['Symbol'].map('{:06}'.format) 

df_stock_price = pd.read_csv('data/df_stock_price.csv', index_col=0)
df_stock_price.index = pd.to_datetime(df_stock_price.index)
df_stock_price
```

```python
display(df_company_info.head())
display(df_stock_price.head())
```

```python
app = dash.Dash()

app.layout = html.Div([

     # Div1
     html.Div(id='Div1',
                 children=[
                     html.H1('Standard Stock Price Graph',
                             style={'textAlign': 'center'}),

                     html.Br(),

                     dcc.Dropdown(id='Div1 Dropdown',
                                  multi=True,
                                  style={'width': '60%',
                                         'fontSize' : 12},
                                  ),
                    # Div1 Date
                    html.Div(id='Div1 Date',
                             children=[
                                 html.H4('Input Start Date & End Date'),
                                 
                                 dcc.Input(id='start_date', 
                                           value='2020-01-01', 
                                           style={'fontSize': 12}),
                                 
                                 dcc.Input(id='end_date', 
                                           value='2021-01-01', 
                                           style={'fontSize': 12}),
                                 
                                 html.Button(id='Div1 Button', 
                                             n_clicks=0, 
                                             children='Submit',
                                             style={'fontSize': 12}),
                                 
                                 ]),
                    # Div1 Graph
                    html.Div(id='Div1 Graph',
                             children = [
                                 dcc.Graph(id='Div1_stock_graph'),
                      ]),
               # Div1 Finish
               ], style={'color': 'green', 'border': 'green solid'}),

     # Div2
     html.Div(id='Div2',
                 children=[
                      # Div2 Layout
                    html.H1('Find Sector',
                              style={'textAlign': 'center'}),
                      
                    html.Br(),
                      
                    html.H4('Input Stock'),
                      
                    dcc.Input(id='Div2 Input', 
                              value='삼성전자', 
                              style={'fontSize': 12}),
                    html.Button(id='Div2 Button', 
                                n_clicks=0, 
                                children='Submit'),
                                ]),

                    html.Br(),

                    # Div2 Sector
                    html.Div(id='Div2 Result',
                             children=['None'],
                             style={'fontSize': 12}),  
                             
                    html.Br(),

                    # Div2 Table                                              
                    html.Div(id='Div2 Table',
                                 children=['None'],
                              ),
               # Div2 Finish
               ], style={'color': 'blue', 'border': 'blue solid'})                 


@app.callback([Output('Div2 Result', 'children'),
               Output('Div2 Table', 'children'),
               Output('Div1 Dropdown', 'options')],
              [Input('Div2 Button', 'n_clicks')],
              [State('Div2 Input', 'value')])
def search_stock_sector(n_clicks, stock_name):
    
    # Div2 Result
    sector = df_company_info[df_company_info['Name'] == stock_name]['Sector'].values[0]

    # Div2 Table
    df_samle = df_company_info[df_company_info['Sector'] == sector][['Symbol', 'Name', 'Industry']]
    data = df_samle.to_dict('records')
    columns = [{'name': i, 'id': i} for i in df_samle.columns]
    tab_res = dt.DataTable(data=data, columns=columns, style_cell={'textAlign': 'center'})

    # Div1 Dropdown
    dropdown_index = df_company_info[df_company_info['Sector'] == sector]['Name']
    drop_res = [{'label': i, 'value': i} for i in dropdown_index]

    return sector, tab_res, drop_res

@app.callback(Output('Div1_stock_graph', 'figure'),
              Input('Div1 Button', 'n_clicks'),
              State('Div1 Dropdown', 'value'),
              State('start_date', 'value'),
              State('end_date', 'value'))
def standard_stock_price_graph(n_clicks, stock_name, start_date, end_date):
    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')

    df_sample = df_stock_price.loc[start:end, stock_name]
    
    data = []

    for col in df_sample:
        trace = go.Scatter(x=df_sample[col].index,
                           y=df_sample[col]/df_sample[col][0],
                           mode='lines',
                           name='{}'.format(col))
        data.append(trace)

    layout = go.Layout(title='Stock Standard Price Graph',
                       yaxis={'tickformat':','})
    fig = go.Figure(data=data, layout=layout)
    fig.update_layout(template='none')

    return fig


if __name__ == '__main__':
     app.run_server(debug=True)
```
